## MatplotLib

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

Significance of "matplotlib inline"

It provides interactivity with the backend in the frontends like the jupyter notebook. It also provides the feature where, the plotting commands below the output cell of the previous plot, will not affect the previous plot, which means it separates different plots.

## Pie Chart

```python
labels='Python', 'C++', 'Ruby', 'Java'
size=[215, 130, 245, 210]
colors=['gold', 'yellowgreen', 'lightcoral','lightskyblue']
explode=(0.4, 0, 0, 0) # explode the first slice
# explode=(0, 0.4, 0, 0)# explode the 2nd slice

plt.pie(size, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')
plt.show

# autopct='%1.1f%%'  -  it shows that pecentage will be shown in
# floating point decimal (here, --.-%)
```

```python
labels='Python', 'C++', 'Ruby', 'Java'
size=[215, 130, 245, 210]
colors=['gold', 'yellowgreen', 'lightcoral','lightskyblue']
#explode=(0.4, 0, 0, 0) # explode the first slice
explode=(0, 0.6, 0, 0)# explode the 2nd slice

plt.pie(size, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')
plt.show
```

```python
labels='Python', 'C++', 'Ruby', 'Java', 'Assembly'
size=[215, 130, 245, 210, 200]
colors=['gold', 'yellowgreen', 'lightcoral','lightskyblue','red']
#explode=(0.4, 0, 0, 0) # explode the first slice
explode=(0, 0, 0, 0, 0.1)# explode the 2nd slice

plt.pie(size, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show

# By default Shadow of a slice is False, to turn on pass a cammand of True
```

## Percentile

- A percentile is a value below which a certain percentage of observation lie.
- 25 percentile = 25% of entire distribution is less than particular value
- Example
- Data set= [2,2,3,4,5,5,5,6,7,8,8,8,8,8,9,9,10,11,12]
- what is the percentile ranking of 10
- percentile ranking of 10 = [(no. of values below 10)/n ] * 100
- n = total no. of elements
- Ans = 16*100/20 = 80%
- 80% of entire distribution is lesss than 10


- what is the percentile ranking of 11
- Ans = 17*100/20 = 85%


- What value exist at percentile ranking of 25%
- Ans: (percentile/100) * (n+1) = 25 * 21 / 100 = 5.25 = it is index position
- Ans: 5 is the value of 25% percentile


- What value exist at percentile ranking of 75%
- Ans: 75 * 21 / 100 = 15.75 = it is index position
- Ans: 9 is the value of 75% percentile

```python
Data= [2,2,3,4,5,5,5,6,7,8,8,8,8,8,9,9,10,11,12]
plt.boxplot(Data,vert=True,patch_artist=True);
```
